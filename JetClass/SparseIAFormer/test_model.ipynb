```python
import sys, torch

from data_handling_pyg import JetClassLightningDataModule
from models_pyg import SparseTransformer

# Use the available HDF5 file(s); reuse the same path if needed
train_path = '../train_00.h5'
val_path = train_path
test_path = train_path

# Small batch for a quick sanity check
dm = JetClassLightningDataModule(
    train_files=train_path,
    val_files=val_path,
    test_files=test_path,
    batch_size=4,
    num_workers=2,
    pin_memory=False,
    persistent_workers=False,
    prefetch_factor=1,
)

dm.setup('fit')
loader = dm.train_dataloader()

# Take one batch
batch = next(iter(loader))

# Build and run the model (CPU)
model = SparseTransformer(in_channels=11, embed_dim=64, edge_dim=6, num_heads=8, num_blocks=2)
model.eval()
with torch.no_grad():
    logits = model(batch.x, batch.edge_index, batch.edge_attr, batch.batch)

print('logits:', logits.shape)
print('labels:', batch.y.shape)
```
